```python
##importamos las librerias
import pandas as pd
```

```python
##leyemos los datoa del csv

df = pd.read_csv('docs/facebook_reviews_disneylandParis_format.csv',sep=',')
df.head()
```

```python
###remplazamos por nombres mas amigables
df.columns = ['ID Usuario','Reseña','Rating','Fecha','Hora Completa','Hora','Dia Semana','Formato Reviwe','Idioma','Mes','Tamaño','Palabras']
df.head()
```

```python
##calculamos promedio de rating por hora
##indicamos el dato de cual lo queremos agrupar primero y luego el dato por el cual queremos crear el promedio
df.groupby(['Hora'])['Rating'].mean()
##se puede obsertar que a las 10 hubo un rating mas alto
```

```python
##calculamos  el rating por hora

#se puede observar que a las 9 es donde hay rating mas sumado

df.groupby(['Hora'])['Rating'].count()
```

```python
##calculamos el tamaño de las reviews por dia de la semana
df.groupby(['Dia Semana'])['Tamaño'].sum()

###donde primero agrupamos con el dia de la semana
##donde segundo calcualmos el tamaño de las reviews
```

```python
#Creando tabla para transformar a español
dia_semana = {'Origen':["Lun","Mar","Mer","Jeu","Ven","Sam","Dim"], 'Dia_Semana':["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]} 
dia_semana = pd.DataFrame(dia_semana)

dia_semana.head()
```

```python
dia_semana
```

```python
##realizamos merge de ambas tablas
df4 = df.merge(dia_semana,left_on='Dia Semana',right_on='Origen',how='left')
df4.head()
```

```python
##calculamos el tamaño de reviews por dia de la semana
df4.groupby(["Dia_Semana"])["Tamaño"].sum()
```

```python
###pivot, no sirve para dar nuevo formato a tablas simples conviertdolas en tablas pivod

df5 = df4.groupby(['Dia_Semana',"Hora"])["Rating"].mean().reset_index()
df5 
```
