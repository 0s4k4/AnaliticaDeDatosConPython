```python
import pandas as pd
```

```python
df = pd.read_csv('../docs/Programas de Universidades.csv',sep='|',encoding='LATIN-1')
df.head()
```

```python
##realizamos el cambio de nombre de columna
df1 = df.copy()
df1 = df1.rename(columns={'NOMBRE':'NOMBRE DE UNIVERSIDAD','CODIGO_ENTIDAD':'CODIGO DE ENTIDAD'})
df1.head()
```

```python
##extraemos la informacion de la universidad cesar vallejo
df1 = df1[df1['NOMBRE DE UNIVERSIDAD'].isin(["UNIVERSIDAD CÉSAR VALLEJO S.A.C."])]
df1.head()
```

```python
##EXTRAEMOS LA LISTA DE LOS PROGRAMAS DE DOCTORADO DEL NUEVO DATAFRAME CREADO
df1.loc[df1['NIVEL_ACADEMICO'] == 'DOCTORADO',"DENOMINACION_PROGRAMA"].unique
df1.head()
```

```python
###REMPLAZAR EL DOCTORADO EN PSICOLOGIA PÓR DOCTORADO EN CIENCIADS DE LA SALUD MENTAL
df1.loc[df1['DENOMINACION_PROGRAMA']=="DOCTORADO EN PSICOLOGIA","DENOMINACION_PROGRAMA"] = "DOCTORADO EN CIENCIAS DE LA SALUD MENTAL"

df1.loc[df1['NIVEL_ACADEMICO']=="DOCTORADO","DENOMINACION_PROGRAMA"].unique()
```

```python
###ELIMINAMOS COLUMNAS 
df1 = df1.drop(columns=["TIPO_AUTORIZACION_PROGRAMA","TIPO_AUTORIZACION_PROGRAMA_LOCAL"])
df1.head()
```

```python
##CREAMOS UNA NUEVA COLUMNA
df1['ACTUALIZADO_POR']='FIORELLA HENRIQUEZ'
df1.head()
```

```python
##convertimos float a string

print(df1["CODIGO_CLASE_PROGRAMA_N2"].dtype)

df1['CODIGO_CLASE_PROGRAMA_N2']=df1['CODIGO_CLASE_PROGRAMA_N2'].astype(str)

print(df1["CODIGO_CLASE_PROGRAMA_N2"].dtype)
```

```python
##utilizamos la funcion lamba apply y lamba
df1["PERIODO_LICENCIAMIENTO_2019"] = df1["PERIODO_LICENCIAMIENTO"].apply(lambda x: x-1)
df1.head()
```
